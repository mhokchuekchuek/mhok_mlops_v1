```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score

import mlflow
import mlflow.sklearn

import numpy as np
import time

# Launch the experiment on mlflow
experiment_name = "electricityconsumption-forecast"
mlflow.set_experiment(experiment_name)

with mlflow.start_run(run_name = "evalute_eiei"):
    # Prepare the data in mumpy (more confortable with that)
    array_inputs_test = pd.DataFrame(pipeline_ja.fit_transform(X_test))
    array_output_test = y_test.values
    # Make the prediction
    tic1 = time.time()
    prediction = model.predict(array_inputs_test)
    duration_prediction = time.time() - tic1

    # Evaluate the model prediction
    metrics = {
    "rmse" : np.sqrt(mean_squared_error(array_output_test, prediction)),
    "mae" : mean_absolute_error(array_output_test, prediction),
    "r2" : r2_score(array_output_test, prediction),
    "duration_prediction" : duration_prediction

    }


    # Log in mlflow (metrics)
    mlflow.log_metrics(metrics)

    # log in mlflow (model)
    mlflow.sklearn.log_model(model, f"model")
```

```python
def Eval_mlflow(X,y, pipeline, model):    
    array_inputs_test = pd.DataFrame(pipeline.fit_transform(X))
    array_output_test = y.values
    # Make the prediction
    tic1 = time.time()
    prediction = model.predict(array_inputs_test)
    duration_prediction = time.time() - tic1

    # Evaluate the model prediction
    metrics = {
    "rmse" : np.sqrt(mean_squared_error(array_output_test, prediction)),
    "mae" : mean_absolute_error(array_output_test, prediction),
    "r2" : r2_score(array_output_test, prediction),
    "duration_prediction" : duration_prediction

    }


    # Log in mlflow (metrics)
    mlflow.log_metrics(metrics)

    # log in mlflow (model)
    mlflow.sklearn.log_model(model, f"model")
```

```python
import mlflow
import mlflow.sklearn

import numpy as np
import time

# Launch the experiment on mlflow
experiment_name = "eval_ngaaaa"
mlflow.set_experiment(experiment_name)

with mlflow.start_run(run_name = "evalute_eieiza", nested = True) as eval_run:
    with mlflow.start_run(run_name = "train_set", nested=True):
        Eval_mlflow(X_train,y_train, pipeline_ja, model)
    with mlflow.start_run(run_name = "test_set", nested=True):
        Eval_mlflow(X_test,y_test, pipeline_ja, model)
```
